```python
import arcpy 
arcpy.env.workspace = r"C:\acgis\project\working\HousingCensus2021\HousingCensus2021.gdb"
input_fc = "Province_21"
```

```python
fields = arcpy.ListFields(input_fc)

# Create a list to store the field names
field_names = []

# Iterate through the fields and add the names to the list
for field in fields:
    field_names.append(field.name)

# Print all the field names separated by a comma
print(", ".join(field_names))
```

```python
#housing suitability percentage
# New field names for calculated percentages
percent_suitable_field = "Percent_Suitable"
percent_not_suitable_field = "Percent_NotSuitable"

# List of field names
field_names = ['Suitable', 'NotSuitabl', 'TotHousiSu']

# Add new fields for calculated percentages
arcpy.management.AddField(input_fc, percent_suitable_field, "DOUBLE")
arcpy.management.AddField(input_fc, percent_not_suitable_field, "DOUBLE")

# Create a cursor to access the attribute table
with arcpy.da.UpdateCursor(input_fc, field_names + [percent_suitable_field, percent_not_suitable_field]) as cursor:
    for row in cursor:
        # Get the field values
        suitable = row[0]
        not_suitable = row[1]
        tot_housing_su = row[2]

        # Check if the field values are not None
        if suitable is not None and not_suitable is not None and tot_housing_su is not None:
            # Calculate the percentages
            if tot_housing_su != 0:
                percent_suitable = (suitable / tot_housing_su) * 100
                percent_not_suitable = (not_suitable / tot_housing_su) * 100
            else:
                percent_suitable = 0
                percent_not_suitable = 0
        else:
            percent_suitable = None
            percent_not_suitable = None

        # Update the fields with the calculated percentages
        row[3] = percent_suitable
        row[4] = percent_not_suitable

        # Update the row in the attribute table
        cursor.updateRow(row)

print("Percentage calculation completed.")
```

```python
#percent of income on shelter
# New field names for calculated percentages
percent_f_30_field = "Percent_F_30"
percent_f_15_field = "Percent_F_15"
percent_f15_30_field = "Percent_F15_30"
percent_f_30_1_field = "Percent_F_30_1"
percent_f30_50_field = "Percent_F30_50"
percent_f_50_field = "Percent_F_50"
percent_f50_100_field = "Percent_F50_100"

# List of field names
field_names = ['F_30', 'F_15', 'F15_30', 'F_30_1', 'F30_50', 'F_50', 'F50_100', 'Cost_incom']

# Add new fields for calculated percentages
arcpy.management.AddField(input_fc, percent_f_30_field, "DOUBLE")
arcpy.management.AddField(input_fc, percent_f_15_field, "DOUBLE")
arcpy.management.AddField(input_fc, percent_f15_30_field, "DOUBLE")
arcpy.management.AddField(input_fc, percent_f_30_1_field, "DOUBLE")
arcpy.management.AddField(input_fc, percent_f30_50_field, "DOUBLE")
arcpy.management.AddField(input_fc, percent_f_50_field, "DOUBLE")
arcpy.management.AddField(input_fc, percent_f50_100_field, "DOUBLE")

# Create a cursor to access the attribute table
with arcpy.da.UpdateCursor(input_fc, field_names + [percent_f_30_field, percent_f_15_field, percent_f15_30_field,
                                                    percent_f_30_1_field, percent_f30_50_field, percent_f_50_field,
                                                    percent_f50_100_field]) as cursor:
    for row in cursor:
        # Get the field values
        f_30 = row[0]
        f_15 = row[1]
        f15_30 = row[2]
        f_30_1 = row[3]
        f30_50 = row[4]
        f_50 = row[5]
        f50_100 = row[6]
        cost_incom = row[7]

        # Check if the cost_incom field value is not None
        if cost_incom is not None:
            # Calculate the percentages
            if cost_incom != 0:
                percent_f_30 = (f_30 / cost_incom) * 100
                percent_f_15 = (f_15 / cost_incom) * 100
                percent_f15_30 = (f15_30 / cost_incom) * 100
                percent_f_30_1 = (f_30_1 / cost_incom) * 100
                percent_f30_50 = (f30_50 / cost_incom) * 100
                percent_f_50 = (f_50 / cost_incom) * 100
                percent_f50_100 = (f50_100 / cost_incom) * 100
            else:
                percent_f_30 = 0
                percent_f_15 = 0
                percent_f15_30 = 0
                percent_f_30_1 = 0
                percent_f30_50 = 0
                percent_f_50 = 0
                percent_f50_100 = 0
        else:
            percent_f_30 = None
            percent_f_15 = None
            percent_f15_30 = None
            percent_f_30_1 = None
            percent_f30_50 = None
            percent_f_50 = None
            percent_f50_100 = None

        # Update the fields with the calculated percentages
        row[8] = percent_f_30
        row[9] = percent_f_15
        row[10] = percent_f15_30
        row[11] = percent_f_30_1
        row[12] = percent_f30_50
        row[13] = percent_f_50
        row[14] = percent_f50_100

        # Update the row in the attribute table
        cursor.updateRow(row)

print("Percentage calculation completed.")
```

```python
#ownership percentages
# New field names for calculated percentages
percent_owner_field = "Percent_Owner"
percent_with_mortgage_field = "Percent_WithMortga"
percent_without_mortgage_field = "Percent_Wo_mortgag"
percent_renter_field = "Percent_Renter"
percent_subsid_hou_field = "Percent_Subsid_hou"
percent_not_subsid_field = "Percent_NotSubsid"
percent_dwell_provi_field = "Percent_DwellProvi"

# List of field names
field_names = ['Owner', 'WithMortga', 'Wo_mortgag', 'Renter', 'Subsid_hou', 'NotSubsid', 'DwellProvi', 'Total']

# Add new fields for calculated percentages
arcpy.management.AddField(input_fc, percent_owner_field, "DOUBLE")
arcpy.management.AddField(input_fc, percent_with_mortgage_field, "DOUBLE")
arcpy.management.AddField(input_fc, percent_without_mortgage_field, "DOUBLE")
arcpy.management.AddField(input_fc, percent_renter_field, "DOUBLE")
arcpy.management.AddField(input_fc, percent_subsid_hou_field, "DOUBLE")
arcpy.management.AddField(input_fc, percent_not_subsid_field, "DOUBLE")
arcpy.management.AddField(input_fc, percent_dwell_provi_field, "DOUBLE")

# Create a cursor to access the attribute table
with arcpy.da.UpdateCursor(input_fc, field_names + [percent_owner_field, percent_with_mortgage_field, percent_without_mortgage_field,
                                                    percent_renter_field, percent_subsid_hou_field, percent_not_subsid_field,
                                                    percent_dwell_provi_field]) as cursor:
    for row in cursor:
        # Get the field values
        owner = row[0]
        with_mortgage = row[1]
        without_mortgage = row[2]
        renter = row[3]
        subsid_hou = row[4]
        not_subsid = row[5]
        dwell_provi = row[6]
        total = row[7]

        # Check if the total field value is not None
        if total is not None:
            # Calculate the percentages
            if total != 0:
                percent_owner = (owner / total) * 100
                percent_with_mortgage = (with_mortgage / total) * 100
                percent_without_mortgage = (without_mortgage / total) * 100
                percent_renter = (renter / total) * 100
                percent_subsid_hou = (subsid_hou / total) * 100
                percent_not_subsid = (not_subsid / total) * 100
                percent_dwell_provi = (dwell_provi / total) * 100
            else:
                percent_owner = 0
                percent_with_mortgage = 0
                percent_without_mortgage = 0
                percent_renter = 0
                percent_subsid_hou = 0
                percent_not_subsid = 0
                percent_dwell_provi = 0
        else:
            percent_owner = None
            percent_with_mortgage = None
            percent_without_mortgage = None
            percent_renter = None
            percent_subsid_hou = None
            percent_not_subsid = None
            percent_dwell_provi = None

        # Update the fields with the calculated percentages
        row[8] = percent_owner
        row[9] = percent_with_mortgage
        row[10] = percent_without_mortgage
        row[11] = percent_renter
        row[12] = percent_subsid_hou
        row[13] = percent_not_subsid
        row[14] = percent_dwell_provi

        # Update the row in the attribute table
        cursor.updateRow(row)

print("Percentage calculation completed.")
```

```python
#maintenance percentages
# New field names for calculated percentages
percent_reg_maint_ne_field = "Percent_RegMaintNe"
percent_min_rep_need_field = "Percent_MinRepNeed"
percent_maj_rep_need_field = "Percent_MajRepNeed"

# List of field names
field_names = ['RegMaintNe', 'MinRepNeed', 'MajRepNeed', 'TotalDwell']

# Add new fields for calculated percentages
arcpy.management.AddField(input_fc, percent_reg_maint_ne_field, "DOUBLE")
arcpy.management.AddField(input_fc, percent_min_rep_need_field, "DOUBLE")
arcpy.management.AddField(input_fc, percent_maj_rep_need_field, "DOUBLE")

# Create a cursor to access the attribute table
with arcpy.da.UpdateCursor(input_fc, field_names + [percent_reg_maint_ne_field, percent_min_rep_need_field, percent_maj_rep_need_field]) as cursor:
    for row in cursor:
        # Get the field values
        reg_maint_ne = row[0]
        min_rep_need = row[1]
        maj_rep_need = row[2]
        total_dwell = row[3]

        # Check if the field values and total_dwell are not None
        if reg_maint_ne is not None and min_rep_need is not None and maj_rep_need is not None and total_dwell is not None:
            # Calculate the percentages
            if total_dwell != 0:
                percent_reg_maint_ne = (reg_maint_ne / total_dwell) * 100
                percent_min_rep_need = (min_rep_need / total_dwell) * 100
                percent_maj_rep_need = (maj_rep_need / total_dwell) * 100
            else:
                percent_reg_maint_ne = 0
                percent_min_rep_need = 0
                percent_maj_rep_need = 0
        else:
            percent_reg_maint_ne = None
            percent_min_rep_need = None
            percent_maj_rep_need = None

        # Update the fields with the calculated percentages
        row[4] = percent_reg_maint_ne
        row[5] = percent_min_rep_need
        row[6] = percent_maj_rep_need

        # Update the row in the attribute table
        cursor.updateRow(row)

print("Percentage calculation completed.")
```
